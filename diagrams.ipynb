```python
import os
import gudhi
import matplotlib.pyplot as plt
gudhi.persistence_graphical_tools._gudhi_matplotlib_use_tex=False
directory = './outputs'
for filename in os.listdir(directory):
    s = directory + '/' + filename
    ax = gudhi.plot_persistence_diagram(persistence_file=s, legend=True)
    ax.set_title(filename[:-9])
    ax.set_aspect("equal")
    plt.show()
```
